# Univariate Visualisation

```python
# import library yang digunakan untuk visualisasi data

import matplotlib.pyplot as plt
```

```python
# Sampel data kota dan jumlah penduduk
cities = ('Bogor', 'Bandung', 'Jakarta', 'Semarang', 'Yogyakarta',
          'Surakarta','Surabaya', 'Medan', 'Makassar')

populations = (45076704, 11626410, 212162757, 19109629, 50819826, 17579085,
               3481, 287750, 785409)
```

## Bar chart

```python
# Membuat bar chart dari data sample
plt.bar(x=cities, height=populations)

plt.show()
```

Dari diagram batang diatas, dapat kita lihat bahwa keterangan nama kota saling bertabrakan

```python
# Solusi pertama dari masalah diatas
plt.bar(x=cities, height=populations)
plt.xticks(rotation=45)

plt.show()
```

```python
# Better solution with horizontal bar chart
plt.barh(y=cities, width=populations)

plt.show()
```

```python
# Import library untuk membuat dataframe dari data sample
import pandas as pd
```

```python
# sintaks kode untuk membuat dataframe
df = pd.DataFrame({'cities': cities, 'populations': populations})
```

```python
# Melakukan pengurutan data berdasarkan nilai populasi
# inplace=True, agar pengurutan data sungguhan terjadi
df.sort_values(by='populations', inplace=True)
```

```python
# Menunjukan dataframe yang sudah dibuat dan diurutkan
display(df)
```

```python
# Melakukan visualisasi dengan data yang sudah di urutkan
plt.barh(y=df["cities"], width=df["populations"])
plt.xlabel("Population (Millions)")
plt.title("Population of Cities in Indonesia")

plt.show()
```

Selain menggunakan library matplotlib, kita juga bisa membuat bar chart menggunakan library seaborn. Untuk membuat bar chart menggunakan library seaborn, kita bisa menggunakan function barplot()

```python
# Import library seaborn sebagai alternatif library matplotlib
import seaborn as sns
```

```python
sns.barplot(y=df["cities"], x=df["populations"], orient="h", color="blue")
# Label untuk sumbu X
plt.xlabel("Population (Millions)")
# Judul grafiknya
plt.title("Population of Cities in Indonesia")
# Menampilkan grafik yang sudah ditentukan datanya, judul dan keterangan sumbu X
plt.show()
```

## Pie Chart

```python
import matplotlib.pyplot as plt

# Sample data untuk pie chart
flavors = ('Chocolate', 'Vanilla', 'Macha', 'Others')
votes = (50, 20, 30, 10)
# Warna untuk masing-masing data
colors = ('#8B4513', '#FFF8DC', '#93C572', '#E67F0D')
# Explode membuat data "cokelat" terpisah dari grafik utama
# Semakin besar nilai explode maka semakin terpisah / jauh
explode = (0.1, 0, 0, 0)

plt.pie(
    x=votes,
    labels=flavors,
    # auto pct memungkinkan kita untuk menampilkan nilai persen menggunakan format string Python
    autopct='%1.1f%%',
    colors=colors,
    explode=explode
)

# Judul dari grafik
plt.title('Favorite Ice Cream Flavors')

plt.show()
```

## Donut Plot

```python
plt.pie(
    x=votes,
    labels=flavors,
    colors=colors,
    # Menentukan seberapa lebar visualisasi tiap data
    wedgeprops = {'width' : 0.4}
)

plt.show()
```

## Histogram

```python
import matplotlib.pyplot as plt
import numpy as np

# Membuat sample acak dari distribusi normal / Gaussian
# Sample data random memiliki jumlah 250 titik data
# dengan nilai mean sebesar 15 dan standard deviation sebesar 5
x = np.random.normal(15, 5, 250)

# Bins = Jumlah data yang ditampilkan
plt.hist(x=x, bins=15)
plt.show()
```

```python
# Membuat histogram menggunakan library seaborn
import seaborn as sns

sns.histplot(x=x, bins=15)
plt.show()
```

```python
# KDE menggambarkan distribusi data, memberikan wawasan terkait bentuk,
# kecenderungan sentral dan penyebaran data
sns.histplot(x=x, bins=15, kde=True)
plt.show()
```

## Box Plot


*  Box plot bisa digunakan untuk mengidentifikasi Outlier yang pada ada data kuantitatif.
*  Kotak merupakan visual dari representasi nilai IQR beserta ambang bawah dan ambang atas dari sebuah data.


```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Sample data random memiliki jumlah 250 titik data
# dengan nilai mean sebesar 15 dan standard deviation sebesar 5
x = np.random.normal(15, 5, 250)
sns.boxplot(x, orient="h")
plt.show()
```

# Bivariate dan Multivariate Visualisation

*  Bivariate Vis. merupakan bentuk visualisasi data untuk menyajikan informasi dari dua variabel dalam suatu dataset.
*  Multivariate Vis. ialah jenis visualisasi data untuk menggambarkan informasi yang terdapat dalam lebih dari 2 variabel.
*  Keduanya sering digunakan untuk mendeskripsikan hubungan antar variabel

## Scatter Plot

```python
import matplotlib.pyplot as plt

# Membuat sample data untuk scatter plot
lemon_diameter = [6.44, 6.87, 7.7, 8.85, 8.15,
                  9.96, 7.21, 10.04, 10.2, 11.06]
lemon_weight = [112.05, 114.58, 116.71, 117.4, 128.93,
                132.93, 138.92, 145.98, 148.44, 152.81]

plt.scatter(x=lemon_diameter, y=lemon_weight)
plt.show()
```

```python
# Menggunaakan library seaborn
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(x=lemon_diameter, y=lemon_weight)

plt.show()
```

```python
# Untuk mempermudah dalam melihat korelasi atau hubungan antar variabel

sns.regplot(x=lemon_diameter, y=lemon_weight)
plt.show()
```

*  Garis biru menunjukan garis korelasi dari hasil regresi
*  Pita di sekitar garis tersebut menunjukan *confidence level* dari hasil regresi tersebut
    *  Semakin kecil bentuk pita yang dihasilkan menandakan tingkat confidence level yang tinggi

Selain itu, kita juga bisa membuat scatter plot untuk menggambarkan hubungan antar tiga variabel. Namun, untuk kebanyakan kasus kita harus menghindari penggunaan plot tiga dimensi karena sulit diamati dan dipaham

## Menggunakan Komponen Visual Tambahan
*  Color

*  Texture
*  Palette

```python
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yahooFinance

# Mengambil data dari yahoo finance tentang saham bank bca
getBBCAinfo = yahooFinance.Ticker("BBCA.JK")
df = getBBCAinfo.history(period="1y") # Jangka waktu data setahun sebelum hari ini.

df_boxplot = df[["Open", "High", "Low", "Close"]]

sns.boxplot(data = df_boxplot, palette = "colorblind")
plt.ylabel('Price', size = 10)

plt.show()
```
