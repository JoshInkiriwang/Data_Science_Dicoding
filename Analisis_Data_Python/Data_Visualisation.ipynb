# Univariate Visualisation

```python
import matplotlib.pyplot as plt
```

```python
cities = ('Bogor', 'Bandung', 'Jakarta', 'Semarang', 'Yogyakarta',
          'Surakarta','Surabaya', 'Medan', 'Makassar')

populations = (45076704, 11626410, 212162757, 19109629, 50819826, 17579085,
               3481, 287750, 785409)
```

## Bar chart

```python
plt.bar(x=cities, height=populations)

plt.show()
```

Dari diagram batang diatas, dapat kita lihat bahwa keterangan nama kota saling bertabrakan

```python
# Solusi pertama dari masalah diatas
plt.bar(x=cities, height=populations)
plt.xticks(rotation=45)

plt.show()
```

```python
plt.barh(y=cities, width=populations)

plt.show()
```

```python
import pandas as pd
```

```python
df = pd.DataFrame({'cities': cities, 'populations': populations})
```

```python
df.sort_values(by='populations', inplace=True)
```

```python
display(df)
```

```python
plt.barh(y=df["cities"], width=df["populations"])
plt.xlabel("Population (Millions)")
plt.title("Population of Cities in Indonesia")

plt.show()
```

Selain menggunakan library matplotlib, kita juga bisa membuat bar chart menggunakan library seaborn. Untuk membuat bar chart menggunakan library seaborn, kita bisa menggunakan function barplot()

```python
import seaborn as sns
```

```python
sns.barplot(y=df["cities"], x=df["populations"], orient="h", color="blue")
plt.xlabel("Population (Millions)")
plt.title("Population of Cities in Indonesia")

plt.show()
```

# Pie Chart

```python
import matplotlib.pyplot as plt

flavors = ('Chocolate', 'Vanilla', 'Macha', 'Others')
votes = (50, 20, 30, 10)
colors = ('#8B4513', '#FFF8DC', '#93C572', '#E67F0D')
explode = (0.1, 0, 0, 0)

plt.pie(
    x=votes,
    labels=flavors,
    autopct='%1.1f%%',
    colors=colors,
    explode=explode
)

plt.title('Favorite Ice Cream Flavors')

plt.show()
```

# Donut Plot

```python
plt.pie(
    x=votes,
    labels=flavors,
    colors=colors,
    wedgeprops = {'width' : 0.4}
)

plt.show()
```

# Histogram

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(15, 5, 250)

plt.hist(x=x, bins=15)
plt.show()
```

```python
import seaborn as sns

sns.histplot(x=x, bins=15)
plt.show()
```

```python
sns.histplot(x=x, bins=15, kde=True)
plt.show()
```

# Box Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

x = np.random.normal(15, 5, 250)
sns.boxplot(x, orient="h")
plt.show()
```
