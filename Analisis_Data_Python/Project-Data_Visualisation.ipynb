### Membersihkan customer_df

#### Menghapus data yang duplikat


```python
# Menghapus data yang duplikat
customer_df.drop_duplicates(inplace=True)

print('Jumlah duplikasi :', customer_df.duplicated().sum())
```

#### Mengisi Missing Value

```python
# Mengambil data yang memiliki missing value
customer_df[customer_df.gender.isna()]
```

Berdasarkan gambar di atas, dapat dilihat bahwa baris data tersebut masih mengandung banyak informasi penting sehingga sayang jika langsung dibuang. Oleh karena itu, pada kasus ini, kita akan menggunakan metode imputation untuk menangani missing value.

```python
# Gunakanlah method value_counts()untuk mengidentifikasi nilai yang dominan.
customer_df.gender.value_counts()
```

Dominan data pada kategori gender yaitu Prefer not to say. Nilai inilah yang selanjutnya akan kita gunakan sebagai pengganti missing value. Proses penggantian ini dapat dilakukan menggunakan method fillna()

```python
customer_df.fillna(value='Prefer not to say', inplace=True)

# cek apakah masih ada kolom data yang NaN
customer_df.isna().sum()
```

#### Mencari inaccurate data

```python
customer_df[customer_df.age == customer_df.age.max()]
```

kita bisa berasumsi bahwa inaccurate value tersebut terjadi karena human error sehingga kelebihan memasukkan nilai nol. Oleh karena itu, gantilah dengan nilai 70. Proses ini dilakukan dengan memanfaatkan method replace()

```python
customer_df.replace({customer_df['age'].max(): 70}, inplace=True)
```

```python
customer_df.describe()
```

```python
customer_df.replace({customer_df['age'].max(): 50}, inplace=True)
```

## Eksplorasi Data order_df

## Eksplorasi Data order_df dan customer_df

Buat kolom baru bernama Status, Aktif ketika customer_id terdeteksi pada kolom order_df dan sebaliknya

```python
customer_id_in_order_df = order_df.customer_id.tolist()
customer_df["status"] = customer_df["customer_id"].apply(lambda x: "Active" if x in customer_id_in_order_df else "Non Active")
customer_df.sample(5)
```

```python
customer_df.groupby(by="status").customer_id.count()
```

Seperti yang terlihat pada pivot table di atas, terdapat cukup banyak pelanggan yang belum pernah melakukan transaksi sebelumnya. Ini tentunya merupakan kabar buruk bagi kita karena hampir 30% pelanggan kita belum pernah melakukan order sebelumnya.

### Join tabel order_df dan customer_df dengan mode left join.

```python
order_customer_df = pd.merge(
    left = order_df,
    right = customer_df,
    how = "left",
    left_on = "customer_id",
    right_on = "customer_id"
)
order_customer_df.head()
```

### Jumlah Order berdasarkan kota

```python
order_customer_df.groupby(by="city").order_id.nunique().sort_values(ascending=False).reset_index().head(10)
```

berdasarkan hasil tersebut, diketahui bahwa kota Jordanside dan New Ava merupakan dua kota yang memiliki jumlah order terbanyak.

### Jumlah Order berdasarkan state

```python
order_customer_df.groupby(by="state").order_id.nunique().sort_values(ascending=False).reset_index().head(10)
```

South Australia merupakan state yang paling banyak melakukan order

### Jumlah order berdasarkan gender

```python
order_customer_df.groupby(by="gender").order_id.nunique().sort_values(ascending=False).reset_index().head(10)
```

kebanyakan order dilakukan oleh pelanggan yang berjenis kelamin prefer not to say

### Jumlah order berdasarkan kelompok usia

```python
order_customer_df["age_group"] = order_customer_df.age.apply(lambda x: "Youth" if x <= 24 else ("Seniors" if x > 64 else "Adults"))
order_customer_df.groupby(by="age_group").order_id.nunique().sort_values(ascending=False)
```

pelanggan yang paling banyak melakukan order berasal dari kelompok usia Adults.

### Menggabungkan table sales_df dan products_df untuk mencari produk yang paling laris

```python
sales_product_df = pd.merge(
    left = sales_df,
    right = product_df,
    how = "left",
    left_on = "product_id",
    right_on = "product_id"
)

sales_product_df.head()
```

```python
sales_product_df.groupby(by="product_type").agg({
    "sales_id": "nunique",
    "quantity_x": "sum",
    "total_price": "sum"
})
```

Tabel diatas menunjukan bahwa **Trousers** merupakan tipe produk yang palng laris. Tetapi jika dilihat dari keuntungan terbesar, **Jacket** merupakan tipe produk yang paling banyak menyumbang revenue perusahaan.

### Membuat pivot table untuk melihat informasi penjualan berdasarkan nama produk

```python
sales_product_df.groupby(by="product_name").agg({
    "sales_id": "nunique",
    "quantity_x": "sum",
    "total_price": "sum"
}).sort_values(by="total_price", ascending=False)
```

Berdasarkan pivot table diatas, diketahui bahwa produk **Denim** merupakan produk yang paling laris dan juga paling banyak menyumbang pendapatan perusahaan
