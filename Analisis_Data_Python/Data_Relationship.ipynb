```python
# Untuk mengidentifikasi hubungan antar feature bertipe numerik,
# kita bisa menggunakan dua parameter yaitu correlation dan covariance.
```

**Correlation**

#### Korelasi ini digambarkan menggunakan nilai dengan rentang -1 hingga 1
### - Nilai Negatif menggambarkan korelasi berlawanan (negative correlation)
### - Nilai Positif menggambarkan korelasi bersesuaian (positive correlation)
### - Nilai 0 artinya tidak memiliki korelasi sama sekali



```python
# Menggunakan library Pandas untuk menghitung korelasi
import pandas as pd

sample_data = {
    'name': ['John', 'Alia', 'Ananya', 'Steve', 'Ben'],
    'age': [24, 22, 23, 25, 28],
    'communication_skill_score': [85, 70, 75, 90, 90],
    'quantitative_skill_score': [80, 90, 80, 75, 70]
}

df = pd.DataFrame(sample_data)

df.corr(numeric_only=True)
```

```python
# Feature age dan communication_skills_score memiliki korelasi cukup bersesuaian
# Feature age dan quantitative_skill_score memikiki korelasi yang sangat berlawanan
```

**Covariance**

#### Untuk mengidentifikasi hubungan antar dua feature dalam sebuah dataset

*  Positive covariance yang menggambarkan dua feature yang berkorelasi positif atau bersesuaian.


*  Negative covariance yang merepresentasikan dua feature yang berkorelasi negatif atau berlawanan.


*  Zero covariance yang menandakan dua feature yang tidak berkorelasi satu sama lain.

```python
# Menggunakan library Pandas untuk menghitung covariance
import pandas as pd

sample_data = {
    'name': ['John', 'Alia', 'Ananya', 'Steve', 'Ben'],
    'age': [24, 22, 23, 25, 28],
    'communication_skill_score': [85, 70, 75, 90, 90],
    'quantitative_skill_score': [80, 90, 80, 75, 70]
}

df = pd.DataFrame(sample_data)

df.cov(numeric_only=True)
```
